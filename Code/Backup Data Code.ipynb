```python
# Get the column name using its index
column_to_rename = df1.columns[0]

# Rename the column
df1.rename(columns={column_to_rename: 'text'}, inplace=True)

# Create a new DataFrame with only the 'text' column
df = df1[['text']].copy()

# Create a new column with a range of numbers
df['No'] = range(1, len(df) + 1)

cols = ['No']  + [col for col in df if col != 'No']
df = df[cols]
df.info()
```

```python
from gensim.models import Word2Vec

# Load the pre-trained Word2Vec model
model = Word2Vec.load('C:\\Users\\dimas\\Downloads\\word2vec-bahasa-indonesia\\model\\idwiki_word2vec.model')
```

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec

# Tokenize the text data
df['tokenized_text'] = df['text'].apply(word_tokenize)

# Get Word2Vec embeddings for each word in the text
df['word2vec_embeddings'] = df['tokenized_text'].apply(lambda x: [model.wv[word] for word in x if word in model.wv])

# Now, each item in 'word2vec_embeddings' is a list of Word2Vec embeddings for the words in the corresponding text
```

```python
import numpy as np

# Function to calculate the mean vector of a list of vectors
def mean_vector(word2vec_embedding_list):
    # Convert the list to a numpy array and calculate the mean along axis 0 (i.e., calculate the mean of each dimension)
    return np.mean(np.array(word2vec_embedding_list), axis=0)

# Apply the function to the 'word2vec_embeddings' column to get a mean vector for each text
df['mean_word2vec_embedding'] = df['word2vec_embeddings'].apply(mean_vector)
```

2 Opsi Cluster

```python
# Filter out texts with no Word2Vec embeddings
df = df[df['word2vec_embeddings'].apply(len) > 0]

# Calculate the mean Word2Vec embedding for each text
df['mean_word2vec_embedding'] = df['word2vec_embeddings'].apply(mean_vector)

# Convert the 'mean_word2vec_embedding' column to a list
X = df['mean_word2vec_embedding'].tolist()

# Initialize the KMeans model
kmeans = KMeans(n_clusters=10, random_state=0)

# Fit the model to your data and predict the cluster labels
df['cluster_labels'] = kmeans.fit_predict(X)
```
